## The place with the highest cases

```python
place_counts
```

```python
health.columns
```

## Top 5 Place and Indicator Category Combinations by Occurrences using seaborn 

```python
places = [f"{place}, {category}" for (place, category) in place_counts.index]
counts = place_counts.values
data = pd.DataFrame({'Place and Indicator Category': places, 'Count': counts})

# Create the bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(y='Place and Indicator Category', x='Count', data=data, hue='Place and Indicator Category', palette='Blues_d', legend=False)


# Add labels and title
plt.xlabel('Count of Occurrences')
plt.ylabel('Place and Indicator Category')
plt.title('Top 5 Place and Indicator Category Combinations by Occurrences');
```

## Trying a new plot known as ggplot

```python
plt.style.available
plt.style.use("ggplot")
```

```python
# Extract "Place" and "Indicator Category" labels

places = [f"{place}, {category}" for (place, category) in place_counts.index]
counts = place_counts.values

# Create a bar plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(places, counts, color='skyblue')

# Add labels and title
ax.set(title='Top 5 Place and Indicator Category Combinations by Occurrences',
       xlabel='Count of Occurrences',
       ylabel='Place and Indicator Category');
```

## Years with highest cases and the type of case

```python
cases = [f"{year}, {category}" for (year, category) in highest_year_death.index]
sum_up = highest_year_death.values
fig, ax = plt.subplots(figsize=(12,6))
ax.barh(cases, sum_up, color="skyblue")
ax.set(title= "Years with highest cases",
       xlabel= "Total",
       ylabel= "years/Indicator Categories");
```
